```python
def constraints(word):
    if 1 <= len(word) <= 100:
        for i in word.lower():
            if 96 < ord(i) < 123:
                continue
            else:
                return False
        return True
    return False
```

```python
def bigger_is_biggest(w):
    if constraints(w):
        word = list(w[::-1].lower())
        for i in range(len(word)):
            for j in range(1, len(word)):
                if word[i] > word[j]:
                    word[i], word[j] = word[j], word[i]
                    return ''.join(word[::-1])
                else:
                    return 'no answer'

    else:
        return 'wrong word'
```

```python
def main_func():
    t = int(input('how many words?: '))
    if 1 <= t <= 100:
        result = [bigger_is_biggest(input(f'word {i + 1}: ').strip()) for i in range(t)]
        for i in result:
            print(i)
    else:
        print('Wrong number. Please enter a number from 1-100')


main_func()
```

# Exercise 3

```python
def bomber_man(row):
    arr = [(list(input().lower())) for i in range(row)]
    destroyed = [i.copy() for i in arr]
    for i in range(len(arr)):
        for j in range(len(destroyed[i])):
            if destroyed[i][j] == 'o' and arr[i][j] == 'o':
                if j > 0:
                    destroyed[i][j - 1] = 'o'
                if i > 0:
                    destroyed[i - 1][j] = 'o'
                if j < len(destroyed[i]) - 1:
                    destroyed[i][j + 1] = 'o'
                if i < len(destroyed) - 1:
                    destroyed[i + 1][j] = 'o'
    return [[i.replace('o', '.') if i == 'o' else i.replace('.', 'O')
             for i in destroyed[j]] for j in range(len(destroyed))]
```

```python
for i in bomber_man(int(input('Enter the number of rows: '))):
    for j in i:
        print(j.strip(), end='')
    print()
```
